```python
x=data[['YearsExperience']]#input
#.values for output in array else come in series
y=data['Salary'].values#output
print(x,'\n',y)
```

```python
#splitting
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)#if train_size=0.8
#The random_state parameter in the train_test_split function is used to control the shuffling process that happens before splitting the data into training and testing sets. Setting random_state to a fixed value (like 0 in your code) ensures that you will get the same split every time you run the code, making your results reproducible.
#If you don't set a value, a random number generator is used, and the split will be different each time.
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
#model.LinearRegression()
```

The main difference between fit (train on given data)and fit_transform (scaling) lies in their purpose and the kind of objects they operate on.

**fit:**

Used for training a machine learning model on the provided data (e.g., x_train, y_train).
Learns the parameters of the model (like coefficients in linear regression) based on the training data.

**fit_transform:**

Typically used with transformers (objects that preprocess data, like scalers or encoders).
It performs two actions:
fit: Learns the necessary parameters from the training data (e.g., mean and standard deviation for a scaler).
transform: Applies the learned transformation to the data (e.g., scaling the data).

```python
ypred=model.predict(x_test)
```

```python
ypred
```

```python
#Error
#from sklearn.metrics import mean_absolute_error
from sklearn  import metrics
import numpy as np
print('MAE: ',metrics.mean_absolute_error(y_test,ypred))
print('MSE: ',metrics.mean_squared_error(y_test,ypred))
print('RMSE: ',np.sqrt(metrics.mean_squared_error(y_test,ypred)))
```
