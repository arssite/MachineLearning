```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Generate synthetic data with non-linear relationships
np.random.seed(0)
X = np.sort(5 * np.random.rand(50, 1), axis=0)
y = np.sin(X).ravel() + 0.1 * (np.random.randn(50))

# Define degrees for polynomial regression models
degrees = [1, 10]

plt.figure(figsize=(12, 6))

# Plot original data
plt.scatter(X, y, color='blue', label='Original data')

for i, degree in enumerate(degrees):
    # Define polynomial regression pipeline
    polynomial_features = PolynomialFeatures(degree=degree)
    linear_regression = LinearRegression()
    pipeline = Pipeline([('polynomial_features', polynomial_features),
                         ('linear_regression', linear_regression)])

    # Fit the model
    pipeline.fit(X, y)

    # Make predictions
    y_pred = pipeline.predict(X)

    # Plot model predictions
    plt.plot(X, y_pred, label='Degree %d' % degree)

plt.title('Underfitting and Overfitting Example')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

We generate synthetic data with non-linear relationships using a sine function with added noise.

We fit two polynomial regression models with different degrees (1 and 10) to the data.

We visualize the original data and the model predictions for both degrees.

________________________________________

The resulting plot will illustrate the difference between underfitting (degree 1) and overfitting (degree 10). You'll see that the degree 1 model is too simplistic and fails to capture the non-linear patterns in the data (underfitting), while the degree 10 model closely follows the training data but generalizes poorly to new data (overfitting).

# **Underfitting and overfitting**

Underfitting and overfitting are two common problems encountered when training machine learning models:

1. **Underfitting**:
   - Underfitting occurs when a model is too simple to capture the underlying structure of the data. It fails to learn the patterns present in the training data and performs poorly on both the training and test datasets.
   - Signs of underfitting include high training and test errors, poor model performance, and inability to capture complex relationships in the data.
   - Underfitting can be caused by using a model that is too simple or lacks the capacity to represent the data adequately, insufficient training, or noisy data.
   - To address underfitting, you can try using a more complex model, increasing the model's capacity, adding more features, reducing regularization, or training the model for longer.

2. **Overfitting**:
   - Overfitting occurs when a model is too complex and captures noise or random fluctuations in the training data instead of the underlying patterns. It performs well on the training data but generalizes poorly to unseen data.
   - Signs of overfitting include low training error but high test error, a large gap between training and test performance, and poor generalization to new data.
   - Overfitting can be caused by using a model that is too complex, training on insufficient data, or training for too many epochs.
   - To address overfitting, you can try using a simpler model, reducing the model's capacity, adding regularization (e.g., L1, L2 regularization), using dropout, collecting more training data, or applying data augmentation techniques.

Balancing between underfitting and overfitting is crucial for building machine learning models that generalize well to unseen data. The goal is to find the right level of model complexity that captures the underlying patterns in the data without being overly simplistic or overly complex. This balance is often referred to as the bias-variance tradeoff.
