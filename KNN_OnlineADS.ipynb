```python
data['Gender']=data['Gender'].map({'Male':0,'Female':1})
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
x=data.iloc[:,:-1].values
y=data.iloc[:,-1]
```

```python
Xt,Xtest,Yt,Ytest=train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
Xt
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
Xscaled=sc.fit_transform(Xt)
Xt
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
```

```python
knc=KNeighborsClassifier(n_neighbors=5,metric='manhattan')
knc.fit(Xscaled,Yt)
train_pred=knc.predict(Xscaled)
a1=metrics.accuracy_score(Yt,train_pred)
print("Training",a1)
```

```python
testscal=sc.fit_transform(Xtest)
test_pred=knc.predict(testscal)
a2=metrics.accuracy_score(Ytest,test_pred)
print("Testing",a2  )
```

```python
#from Confusion Metrics
#test
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Ytest,test_pred)
cm
```
